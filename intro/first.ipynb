```python
import pandas as pd

df = pd.read_csv("iris.csv")

print("Dataset shape:", df.shape)

print("Dataset head:")
print(df.head())


print("Dataset tail:")
print(df.tail())
```

## 1-9. Розуміння даних та приклади використання pandas (з коментарями українською)

```python
# 1) Базові перегляди та метадані: head, tail, sample, shape, columns, index, dtypes, info
print('head:', df.head())
print('tail:', df.tail())
print('sample (5):', df.sample(5, random_state=1))
print('shape:', df.shape)
print('columns:', df.columns.tolist())
print('index (first 10):', df.index[:10].tolist())
print('dtypes:', df.dtypes)
print('info:')
df.info()
```

```python
# 2) Створити нову колонку: сума sepal_length + petal_length
if 'sepal_length' in df.columns and 'petal_length' in df.columns:
    df['sum_sepal_petal'] = df['sepal_length'] + df['petal_length']
    print(df[['sepal_length','petal_length','sum_sepal_petal']].head())
else:
    print('Не знайдено потрібних колонок для прикладу сумування')
```

### Indexing, Selecting & Assigning

```python
# Вибір декількох колонок та рядків
print('Декілька колонок:', df[['sepal.length','variety']].head())
# loc (label-based) — тут індекс числовий, покажемо по булевому масці
print('loc: sepal.length > 5 (перші 5):', df.loc[df['sepal.length']>5, ['sepal.length','variety']].head())
# iloc (position-based): перші 3 рядки, перші 2 колонки
print('iloc: first 3 rows, first 2 cols:', df.iloc[:3, :2])
# Видалити непотрібну колонку (не inplace, просто показати результат)
print('drop column sepal_width (preview):', df.drop(columns=['sepal.width']).head())
# Conditional selection
print('Conditional selection (sepal.length between 5 and 6):', df[(df['sepal.length']>5) & (df['sepal.length']<6)].head())
# set_index (тимчасово) за species
df_idx = df.set_index('variety')
print('set_index preview (index becomes species):', df_idx.head())
```

### Summary functions and map

```python
# value_counts, unique, sum, mean, map
print('value_counts species:', df['variety'].value_counts())
print('unique species:', df['variety'].unique())
print('sum sepal_length:', df['sepal.length'].sum())
print('mean petal_length:', df['petal.length'].mean())
# map: перетворимо variety у короткий код
mapping = {s: i for i, s in enumerate(df['variety'].unique(), 1)}
print('mapping example:', mapping)
df['species_code'] = df['variety'].map(mapping)
print(df[['variety','species_code']].head())
# map з lambda — наприклад класифікуємо sepal.length > 5 як 'long'/'short'
df['sepal_len_cat'] = df['sepal.length'].map(lambda x: 'long' if x>5 else 'short')
print(df[['sepal.length','sepal_len_cat']].head())
```

### Grouping and sorting

```python
# Сортування: зростання/спадання
print('sort sepal.length asc (top 5):', df.sort_values('sepal.length', ascending=True).head())
print('sort sepal.length desc (top 5):', df.sort_values('sepal.length', ascending=False).head())
# Групування по одній колонці (variety) — різні агрегації
grp1 = df.groupby('variety')['sepal.length'].agg(['count','mean','max']).round(2)
print('Group by variety (count, mean, max):', grp1)
# Групування по двох колонках — додаємо sepal_len_cat
grp2 = df.groupby(['variety','sepal_len_cat'])['sepal.length'].agg(['count','mean']).round(2)
print('Group by variety + sepal_len_cat:', grp2)
# Агрегувати по кількох параметрах на одній групі (agg с словником)
agg_multi = df.groupby('variety').agg({'sepal.length':['mean','min','max'],'petal.length':['mean','sum']}).round(2)
print('Multiple agg:', agg_multi)
# Фільтрація по значенню та отримати рядки
print('Rows where sepal.length > 7:', df[df['sepal.length']>7])
# Приклад query
print('Query example (sepal.length>5 and variety=="versicolor"):', df.query('`sepal.length`>5 and `variety`=="versicolor"').head())
```
